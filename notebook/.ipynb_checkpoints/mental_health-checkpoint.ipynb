Spliting data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

other preprocessing tasks such as removing unnecessary data, working with categorical data, feature engineering, and feature selection should be performed after splitting the data into training and test sets. This ensures that the test set remains independent and unbiased.

 if you want to fill missing values with the mean of a column, you should calculate the mean using only the training data and then use that value to fill missing values in both the training and test sets. This way, you avoid any information leakage from the test set into the training data1.

```python
df_features=df.drop(columns=["Anxiety","Depression","Insomnia","OCD"],axis=1)
df_targets=np.column_stack((df["Anxiety"],df["Depression"],df["Insomnia"],df["OCD"]))
```

```python
df_features
```

```python
df_targets
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df_features,df_targets,test_size=0.15,stratify=df["Fav genre"],random_state=19)
```

```python
# X_train.info()
```

```python
# separating cat and num columns
categorical_cols=X_train[X_train.columns[X_train.dtypes=="object"]].columns
numerical_cols=X_train[X_train.columns[X_train.dtypes!="object"]].columns
```

```python
# X_train[categorical_cols[0]].mode()[0]
```

```python
# filling for categorical values with most frequent value and the value came is apply to null value of test data
for cat in categorical_cols:
    imputer_value=X_train[cat].mode()[0] #values from train data only
    print(cat,imputer_value)
    X_train[cat]=X_train[cat].fillna(imputer_value)
    X_test[cat]=X_test[cat].fillna(imputer_value)
```

```python
# similar for numerical data
for num in numerical_cols:
    imputer_num_value=X_train[num].mean() #values from train data only
    print(num,imputer_num_value)
    X_train[num]=X_train[num].fillna(imputer_num_value)
    X_test[num]= X_test[num].fillna(imputer_num_value)
    
```

```python
print(X_train.isna().sum(),X_test.isna().sum())
```

```python
X_train
```

```python
# using onehotencoder
# If you perform the encoding before the split, it will lead to data leakage (train-test contamination). In this sense, you will introduce new data (integers of Label Encoders) and use it for your models thus it will affect the end predictions results (good validation scores but poor in deployment).
```

```python
from sklearn.preprocessing import OneHotEncoder
# fit to train data and transform for test data 
```

```python
encoder = OneHotEncoder(dtype="int32")
encoder.fit(X_train[categorical_cols])
```

```python
one_hot = pd.DataFrame(encoder.transform(X_train[categorical_cols]).toarray(), columns=encoder.get_feature_names_out(categorical_cols)).reset_index(drop=True)
# Concatenate the one-hot encoded DataFrame with the original DataFrame
Xnew_train = pd.concat([X_train.drop(columns=categorical_cols).reset_index(drop=True), one_hot], axis=1)
```

```python
Xnew_train
```

```python
one_hot_test = pd.DataFrame(encoder.transform(X_test[categorical_cols]).toarray(), columns=encoder.get_feature_names_out(categorical_cols)).reset_index(drop=True)
one_hot_test
# Concatenate the one-hot encoded DataFrame with the original DataFrame
Xnew_test = pd.concat([X_test.drop(columns=categorical_cols).reset_index(drop=True), one_hot_test], axis=1)
```
