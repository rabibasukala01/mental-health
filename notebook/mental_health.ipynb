other preprocessing tasks such as removing unnecessary data, working with categorical data, feature engineering, and feature selection should be performed after splitting the data into training and test sets. This ensures that the test set remains independent and unbiased.

 if you want to fill missing values with the mean of a column, you should calculate the mean using only the training data and then use that value to fill missing values in both the training and test sets. This way, you avoid any information leakage from the test set into the training data1.

```python
df_features=df.drop("mental_scale",axis=1)
df_target=df["mental_scale"]
```

```python
# df_features
# df_target
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df_features,df_target,test_size=0.17,stratify=df["Fav genre"],random_state=19)
```

```python
X_train.shape
```

```python
# separating cat and num columns
categorical_cols=X_train[X_train.columns[X_train.dtypes=="object"]].columns
numerical_cols=X_train[X_train.columns[X_train.dtypes!="object"]].columns
```

```python
# X_train[categorical_cols[0]].mode()[0]
```

# handling null values

```python
for cat in categorical_cols:
    imputer_value=X_train[cat].mode()[0] #values from train data only [frequent data]
    print(cat,imputer_value)
    X_train[cat]=X_train[cat].fillna(imputer_value)
    X_test[cat]=X_test[cat].fillna(imputer_value)
```

```python
# similar for numerical data
for num in numerical_cols:
    imputer_num_value=X_train[num].mean() #values from train data only
    print(num,imputer_num_value)
    X_train[num]=X_train[num].fillna(imputer_num_value)
    X_test[num]= X_test[num].fillna(imputer_num_value)
```

```python
print(Y_test.isna().sum(),Y_train.isna().sum())
# no null in targets
```

```python
# print(X_train.isna().sum(),X_test.isna().sum())
# handled all null values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

ohe = OneHotEncoder(dtype="int32",sparse_output=False)
ss=StandardScaler()
```

```python
transformer=ColumnTransformer(transformers=[
    ('ohe',ohe,categorical_cols),
    ('ss',ss,numerical_cols)
],remainder="passthrough")
```

```python
Xnew_train=transformer.fit_transform(X_train)
```

```python
Xnew_train.shape,X_train.shape
```

```python
Xnew_test=transformer.transform(X_test)
```

```python
Xnew_test.shape,X_test.shape
```

```python
# saving transformer
import pickle
with open('transformer.pkl','wb') as f:
    pickle.dump(transformer,f)
```
